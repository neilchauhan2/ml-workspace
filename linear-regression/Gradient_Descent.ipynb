```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate data
# function => y = (x - 5)^2 + 3
X = np.arange(10)
Y = (X-5)**2 + 3

print(X,Y)
```

```python
# Visualize
plt.plot(X,Y)
plt.show()
```

```python
# Gradient Descent
# initialize x with any random value

x = 0
lr = 0.1
plt.plot(X,Y)
for i in range(30):
  # differentiate the function to compute the gradient
  grad = 2*(x-5)
  x = x - lr * grad
  y = (x - 5)**2 + 3
  plt.scatter(x,y)
  print(x)

plt.show()
```
